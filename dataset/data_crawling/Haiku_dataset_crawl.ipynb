```python
# Wirtten by ZW.Lin, 09/20/2019

import urllib
import chardet
import re
import requests
import codecs

from bs4 import BeautifulSoup
from urllib import request
from newspaper import Article
```

```python
# Get all urls on http://www.tempslibres.org/tl/tlphp/dbauteursl.php?lang=en&lg=e

src_url = 'http://www.tempslibres.org/tl/tlphp/dbauteursl.php?lang=en&lg=e'
req = request.Request(src_url)
response = request.urlopen(req)
response = response.read().decode('gbk', 'replace') # open url, decode as gbk
#print(response)
soup = BeautifulSoup(response, 'lxml')
urls = []
for tag in soup.find_all('td', class_='liensurl'): # find <td class="liensurl">
    tmp = tag.a  # parse <a href=...>
    if tmp is not None:
        print(tmp.get('href'))
        urls.append('http://www.tempslibres.org/tl/tlphp/'+tmp.get('href'))
for url in urls:
    print(url+'\n')
print(len(urls))
```

```python
# Crawl all content on urls

Haikus = []

for url in urls:
    try:
        a = Article(url, language='en')
        a.download()
        a.parse()
        print(a.text)
        Haikus.append(a.text)
    except Exception:
        continue

output = codecs.open('E:/WPI/19 fall/CS 534 - ARTIFICIAL INTELLIGENCE/dataset/Haikus.txt','w',encoding='utf-8')
for Haiku in Haikus:
    output.write(Haiku)
    output.write('\n')
output.close()
#     print(Haiku+'\n')
print(len(Haikus))
```
